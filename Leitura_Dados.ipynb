```python
import scipy.io as sc
import netCDF4 as nc
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
data4 = nc.Dataset('data2015_4dias.nc4','r')
data4
```

```python
T1 = data4.variables['water_temp'][0,0,:,:]
T2 = data4.variables['water_temp'][21,0,:,:]
V = data4.variables['water_v'][0,0,:,:]
U = data4.variables['water_u'][0,0,:,:]
lon = data4.variables['lon'][:]
lat = data4.variables['lat'] [:]
LON,LAT = np.meshgrid(lon,lat)

```

```python

fig = plt.figure(figsize=(12,8))


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
plt.contourf(LON,LAT,T1)
plt.colorbar()
plt.title('Temperatura da Superficie do mar')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```

```python

fig = plt.figure(figsize=(12,8))


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
plt.contourf(LON,LAT,T2)
plt.colorbar()
plt.title('Temperatura da Superficie do mar')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```

```python

fig = plt.figure(figsize=(12,8))


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
plt.contourf(LON,LAT,T1-T2)
plt.colorbar()
plt.title('Temperatura da Superficie do mar')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```

```python

fig = plt.figure(figsize=(12,8))


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
plt.pcolormesh(LON,LAT,T1,cmap='jet')
plt.colorbar()
plt.title('Temperatura da Superficie do mar')
plt.xlim(-70,-30)
plt.ylim(-60,-30)
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```

```python
S1 = data4.variables['salinity'][0,0,:,:]
S2 = data4.variables['salinity'][21,0,:,:]
```

```python
from matplotlib.cm import ScalarMappable
fig = plt.figure(figsize=(12,8))


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
qcs=plt.pcolormesh(LON,LAT,S1,vmin=29,vmax=39,cmap='jet',transform=ccrs.PlateCarree())
plt.colorbar(   ScalarMappable(norm=qcs.norm, cmap=qcs.cmap),ticks=range(29, 39,1))
plt.title('Salinidade da Superficie do mar')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```
